```python
years = list(range(1980,2014))
print(years)
```

```python
canada.loc['Japan',years].plot(kind='area',title='Japan Immigration to Canada',
    figsize=(15,5))
```

```python
countries = ['Cambodia','Japan','France']
canada.loc[countries,years]
```

```python
canada.loc[countries,years].T.plot(kind='line',title= f'{",".join(countries)}',figsize=(15,5))
plt.savefig('Compairing 3-countries.png')
```

```python
countries = ['Cambodia','Japan','France','India']
canada.loc[countries,years]
```

```python
canada.loc[countries,years].T.plot(kind='area',title= f'{",".join(countries)}',figsize=(15,5))
plt.savefig('Compairing 4-countries.png')
```
